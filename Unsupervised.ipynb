```python
import matplotlib.pyplot as plt

def elbow_plot(X, k_range=(1, 11)):
    inertias = []
    K = range(*k_range)
    
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(X)
        inertias.append(kmeans.inertia_)
    
    plt.figure(figsize=(8, 5))
    plt.plot(K, inertias, 'bo-')
    plt.xlabel('k (number of clusters)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal k')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return np.array(K), np.array(inertias)

k_values, inertias = elbow_plot(X)
```

```python
for k, inertia in zip(k_values, inertias):
    print(f"k={k}: inertia={inertia:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def silhouette_plot(X, k_range=(2, 11)):
    sil_scores = []
    K = list(range(*k_range))
    
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(X)
        score = silhouette_score(X, kmeans.labels_)
        sil_scores.append(score)
        print(f"k={k}: silhouette={score:.3f}")
    
    plt.figure(figsize=(8, 5))
    plt.plot(K, sil_scores, 'ro-', linewidth=2, markersize=8)
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Analysis for Optimal k')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    best_k = K[np.argmax(sil_scores)]
    print(f"\nBest k: {best_k} (silhouette score: {max(sil_scores):.3f})")
    
    return K, sil_scores

k_values, scores = silhouette_plot(X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

def plot_gmm_aic_bic(X, k_min=1, k_max=10, covariance_type='full', random_state=42):
    ks = list(range(k_min, k_max + 1))
    aics, bics = [], []

    for k in ks:
        gm = GaussianMixture(
            n_components=k,
            covariance_type=covariance_type,
            n_init=10,
            random_state=random_state
        )
        gm.fit(X)
        aics.append(gm.aic(X))
        bics.append(gm.bic(X))

    aics = np.array(aics)
    bics = np.array(bics)

    plt.figure(figsize=(8, 5))
    plt.plot(ks, aics, 'o-', label='AIC')
    plt.plot(ks, bics, 's-', label='BIC')
    plt.xlabel('Number of components (k)')
    plt.ylabel('AIC / BIC')
    plt.title('GMM Model Selection via AIC and BIC')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Find best k
    best_k_aic = ks[np.argmin(aics)]
    best_k_bic = ks[np.argmin(bics)]
    print(f"Best k by AIC: {best_k_aic}, AIC={aics.min():.2f}")
    print(f"Best k by BIC: {best_k_bic}, BIC={bics.min():.2f}")

    return ks, aics, bics, best_k_aic, best_k_bic

ks, aics, bics, best_k_aic, best_k_bic = plot_gmm_aic_bic(X, k_min=1, k_max=10)


# we got best size as 3 which is already we worked out
```

```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
import numpy as np

faces = fetch_olivetti_faces()
X = faces.data      
y = faces.target
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.25, 
    stratify=y_train_full, random_state=42
)

print("Shapes:")
print("Train:", X_train.shape, "Val:", X_val.shape, "Test:", X_test.shape)
print("Distinct people in train:", len(np.unique(y_train)))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def kmeans_silhouette_scan(X, k_range, random_state=42):
    sil_scores = []
    for k in k_range:
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=random_state)
        labels = kmeans.fit_predict(X)
        score = silhouette_score(X, labels)
        sil_scores.append(score)
        print(f"k={k}: silhouette={score:.3f}")
    return np.array(sil_scores)

k_range = range(5, 41, 5)
sil_scores = kmeans_silhouette_scan(X_train, k_range)

best_k = k_range[np.argmax(sil_scores)]
print("Best k by silhouette:", best_k)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(list(k_range), sil_scores, 'o-')
plt.xlabel("k (number of clusters)")
plt.ylabel("Silhouette score")
plt.title("K-Means Silhouette on Olivetti Faces")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
k = best_k
kmeans = KMeans(n_clusters=k, n_init=20, random_state=42)
cluster_labels = kmeans.fit_predict(X_train)

```

```python
def plot_cluster_faces(X, labels, cluster_id, n_faces=10, img_shape=(64, 64)):

    idx = np.where(labels == cluster_id)[0]
    n_show = min(len(idx), n_faces)
    if n_show == 0:
        print(f"No faces in cluster {cluster_id}")
        return
    
    plt.figure(figsize=(10, 2))
    for i in range(n_show):
        ax = plt.subplot(1, n_show, i+1)
        face = X[idx[i]].reshape(img_shape)
        plt.imshow(face, cmap="gray")
        plt.axis("off")
    plt.suptitle(f"Cluster {cluster_id} (showing {n_show} faces)")
    plt.tight_layout()
    plt.show()

for c in range(min(5, k)):
    plot_cluster_faces(X_train, cluster_labels, c, n_faces=10)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

base_clf = LinearSVC(random_state=42)
base_clf.fit(X_train, y_train)

y_val_pred = base_clf.predict(X_val)
base_val_acc = accuracy_score(y_val, y_val_pred)
print(f"Baseline validation accuracy: {base_val_acc:.3f}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression

def kmeans_features(X_train, X_val, y_train, y_val, k_values):
    results = []

    for k in k_values:
        print(f"\n== k = {k} clusters ===")

        kmeans = KMeans(n_clusters=k, n_init = 10, random_state = 42)
        kmeans.fit(X_train)

        X_train_k = kmeans.transform(X_train)
        X_val_k = kmeans.transform(X_val)

        clf_k = LogisticRegression(max_iter = 1000, multi_class = 'multinomial', random_state = 42)
        clf_k.fit(X_train_k, y_train)

        y_val_pred_k = clf_k.predict(X_val_k)
        acc = accuracy_score(y_val, y_val_pred_k)
        print(f"Validation accuracy with K-Means features (k={k}): {acc:.3f}")
        
        results.append((k, acc))
        
    return results

k_values = [10, 20, 40, 80, 120, 160]
km_results = kmeans_features(X_train, X_val, y_train, y_val, k_values)

best_k_km, best_acc_km = max(km_results, key=lambda x: x[1])
print(f"\nBest k for K-Means-only features: k={best_k_km}, val_acc={best_acc_km:.3f}")
```

```python
import numpy as np

def kmeans_append_features(X_train, X_val, y_train, y_val, k_values):
    results = []
    for k in k_values:
        print(f"\n=== k = {k} clusters (append) ===")
        
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
        kmeans.fit(X_train)
        
        X_train_k = kmeans.transform(X_train)
        X_val_k   = kmeans.transform(X_val)
        
        # Concatenate original + K-Means features
        X_train_aug = np.hstack([X_train, X_train_k])
        X_val_aug   = np.hstack([X_val,   X_val_k])
        
        clf = LinearSVC(random_state=42)
        clf.fit(X_train_aug, y_train)
        
        y_val_pred = clf.predict(X_val_aug)
        acc = accuracy_score(y_val, y_val_pred)
        print(f"Validation accuracy with appended features (k={k}): {acc:.3f}")
        
        results.append((k, acc))
    return results

k_values_append = [10, 20, 40, 80, 120, 160]
append_results = kmeans_append_features(X_train, X_val, y_train, y_val, k_values_append)

best_k_app, best_acc_app = max(append_results, key=lambda x: x[1])
print(f"\nBest k for appended features: k={best_k_app}, val_acc={best_acc_app:.3f}")
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

faces = fetch_olivetti_faces()
X = faces.data
h, w = faces.images.shape[1:]

pca = PCA(0.99, whiten = False, random_state = 42)
X_pca = pca.fit_transform(X)
print("Original dim:", X.shape[1], "-> PCA dim:", X_pca.shape[1] )

n_components = 40
gm = GaussianMixture(n_components = n_components,
                     covariance_type = 'full',
                     n_init = 5,
                     random_state = 42)

gm.fit(X_pca)
```

```python
n_samples = 10
X_pca_samples, _ = gm.sample(n_samples=n_samples)

X_samples = pca.inverse_transform(X_pca_samples) 

plt.figure(figsize=(10, 2))
for i in range(n_samples):
    ax = plt.subplot(1, n_samples, i+1)
    ax.imshow(X_samples[i].reshape(h, w), cmap='gray')
    ax.axis('off')
plt.suptitle('Sampled faces from GMM in PCA space')
plt.tight_layout()
plt.show()
```

```python
from skimage.transform import rotate
from skimage.util import random_noise

def make_anomalies(X, n=20, angle=45, dark_factor=0.3, flip=True):

    rng = np.random.RandomState(42)
    idx = rng.choice(len(X), size=n, replace=False)
    X_sel = X[idx].reshape(-1, h, w)

    anomalies = []
    for img in X_sel:
        
        mod = img.copy()
        mod = rotate(mod, angle=angle, mode='edge')
        if flip:
            mod = np.fliplr(mod)
        mod = np.clip(mod * dark_factor, 0, 1)
        anomalies.append(mod.reshape(-1))
    return X[idx], np.array(anomalies)

# Select some normal and anomaliy samples
X_normal, X_anom = make_anomalies(X, n=40, angle=45, dark_factor=0.3, flip=True)

X_normal_pca = pca.transform(X_normal)
X_anom_pca   = pca.transform(X_anom)

# Compute log-likelihood scores
normal_scores = gm.score_samples(X_normal_pca)
anom_scores   = gm.score_samples(X_anom_pca)
```

```python
print("Normal scores: mean={:.2f}, std={:.2f}".format(normal_scores.mean(), normal_scores.std()))
print("Anomaly scores: mean={:.2f}, std={:.2f}".format(anom_scores.mean(), anom_scores.std()))

plt.figure(figsize=(8,4))
plt.hist(normal_scores, bins=15, alpha=0.6, label='Normal')
plt.hist(anom_scores, bins=15, alpha=0.6, label='Anomalies')
plt.xlabel('Log-likelihood (GMM score_samples)')
plt.ylabel('Count')
plt.title('GMM scores: normal vs anomalies')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
n_show = 5
plt.figure(figsize=(10, 4))
for i in range(n_show):
    ax = plt.subplot(2, n_show, i+1)
    ax.imshow(X_normal[i].reshape(h,w), cmap='gray')
    ax.axis('off')
    ax.set_title(f"N:{normal_scores[i]:.1f}")

for i in range(n_show):
    ax = plt.subplot(2, n_show, n_show+i+1)
    ax.imshow(X_anom[i].reshape(h,w), cmap='gray')
    ax.axis('off')
    ax.set_title(f"A:{anom_scores[i]:.1f}")

plt.suptitle("Top: normal | Bottom: anomalies (rotated/flipped/darkened)")
plt.tight_layout()
plt.show()


# Now the model can reject any image which isnt matched or slightly blurry based on original images this will help to achieve zero false positive 
```
